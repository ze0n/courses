```python
# Read data from a csv
df = pd.read_csv('../data/g2_param_space_scan.csv', header=0)
df.sample(10)
```

```python
def draw3D(df, cols, fixed):
    sdf = df
    for key in fixed:
        sdf = sdf[sdf[key] == fixed[key]]
    sdf = sdf[cols]
    
    dt = sdf.pivot_table(values=cols[2], columns=cols[0], index=cols[1]).as_matrix()
    
    data = [ go.Surface( z=dt ) ]
    layout = go.Layout(
        title='Calibration surface ' + ",".join(map(lambda x: x + "=" + str(fixed[x]), fixed)),
        autosize=False,
        width=900,
        height=900,
        margin=dict(
            l=65,
            r=50,
            b=65,
            t=90
        )
    )
    fig = go.Figure(data=data, layout=layout)
    return py.iplot(fig, filename='elevations-3d-surface')
```

```python
draw3D(df, ["a_i", "nu_i", "Perf"], {"b_i": 2, "sigma_i":2, "rho_i" : 7})
```

```python
# Read data from a csv
traj_df = pd.read_csv('../data/optimization_trajectory.csv', header=0)
traj_df["Perf"] = 0
traj_df.head(10)
```

```python
def draw3DScatter(df, traj, cols, fixed):
    sdf = df
    for key in fixed:
        sdf = sdf[sdf[key] == fixed[key]]
    sdf = sdf[cols]
    
    trace1 = go.Scatter3d(
        x=sdf[cols[0]],
        y=sdf[cols[1]],
        z=sdf[cols[2]],
        mode='markers',
        marker=dict(
            size=5,
            line=dict(
                color='rgba(217, 217, 217, 0.14)',
                width=0.5
            ),
            opacity=0.8
        )
    )
    
    trace2 = go.Scatter3d(
        x=traj[cols[0]],
        y=traj[cols[1]],
        z=traj[cols[2]],
        mode='markers',
        marker=dict(
            size=5,
            line=dict(
                color='rgba(217, 217, 217, 0.14)',
                width=0.5
            ),
            colorscale='Reds',
            color=traj.index,

            opacity=0.8
        ),
        line=dict(
            color='#1f77b4',
            width=3
        )
    )

    data = [trace1, trace2]
    layout = go.Layout(
        title='Calibration surface ' + ",".join(map(lambda x: x + "=" + str(fixed[x]), fixed)),
        autosize=False,
        width=900,
        height=900,
        margin=dict(
            l=0,
            r=0,
            b=0,
            t=0
        )
    )
    fig = go.Figure(data=data, layout=layout)
    return py.iplot(fig, filename='simple-3d-scatter')
```

```python
draw3DScatter(df, traj_df, ["a", "nu", "Perf"], {"b_i":2, "sigma_i":9, "rho_i" : 3})
```
